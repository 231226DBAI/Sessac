```python

#!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 연설문을 가져옴 (위의 기후 연설문)
speech = '''
Title: Our Responsibility for Climate Change

Respected audience,

Today, I would like to talk about one of the most pressing issues we face—climate change. Climate change is impacting our actions and lifestyles significantly, demanding urgent responses.

We are already witnessing extreme climate events on Earth, leading to global warming and resulting in severe weather conditions, rising sea levels, and ecological changes. This endangers numerous species, including our own, and makes life more challenging for all of us.

We bear responsibility for this issue. Reducing our carbon footprint in our daily lives is not just a choice; it is an obligation. We must consider the environmental impact in our choices, whether it's using cars, consuming energy, or selecting products.

Governments and businesses also play pivotal roles. They must contribute by developing sustainable energy sources, adopting eco-friendly technologies, and actively participating in efforts to protect our planet. Furthermore, there is a need for increased education and awareness about climate change.

However, all these efforts would be in vain without our collective cooperation. We must work together, considering the serious consequences our actions have on future generations. Our choices today will shape the future.

Respected audience, in this crucial moment where our actions determine the Earth's future, what choices will we make? Understanding our responsibility for the climate crisis, we need to join forces to preserve our planet. The time to act is now.

Thank you.
'''

# 공백을 기준으로 단어 추출
words = speech.split()

# 3글자 이상의 단어만 선택
long_words = [word.strip('.,?!') for word in words if len(word.strip('.,?!')) >= 3]

# 빈도 계산
word_counts = Counter(long_words)

# WordCloud 생성
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

# 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
