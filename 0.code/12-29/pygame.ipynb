```python
import pygame
import sys
import random

# Pygame 초기화
pygame.init()

# 화면 설정
width, height = 800, 600
grid_size = 20  # 격자 크기
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption("지렁이 게임")

# 이미지 불러오기 및 리사이즈
snake_image = pygame.image.load("snake.png")
snake_image = pygame.transform.scale(snake_image, (grid_size, grid_size))  # 격자 크기로 리사이즈

food_image = pygame.image.load("food.png")
food_image = pygame.transform.scale(food_image, (grid_size, grid_size))  # 격자 크기로 리사이즈

# 색깔 정의
white = (255, 255, 255)
gray = (200, 200, 200)

# 지렁이 초기 위치 및 속도
snake_size = grid_size
snake_speed = 10  # 초기 속도
snake_x = width // 2
snake_y = height // 2
snake_x_change = 0
snake_y_change = 0

# 먹이 초기 위치
food_size = grid_size
food_x = round(random.randrange(0, width - food_size) / grid_size) * grid_size
food_y = round(random.randrange(0, height - food_size) / grid_size) * grid_size

# 초기 지렁이 길이 및 꼬리 위치 저장
snake_length = 1
snake_list = []

# 게임 루프
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                snake_x_change = -snake_size
                snake_y_change = 0
            elif event.key == pygame.K_RIGHT:
                snake_x_change = snake_size
                snake_y_change = 0
            elif event.key == pygame.K_UP:
                snake_y_change = -snake_size
                snake_x_change = 0
            elif event.key == pygame.K_DOWN:
                snake_y_change = snake_size
                snake_x_change = 0

    # 지렁이 이동
    snake_x += snake_x_change
    snake_y += snake_y_change

    # 벽에 부딪혔을 때 처리
    if snake_x >= width or snake_x < 0 or snake_y >= height or snake_y < 0:
        pygame.quit()
        sys.exit()

    # 먹이 먹었을 때 처리
    if snake_x == food_x and snake_y == food_y:
        food_x = round(random.randrange(0, width - food_size) / grid_size) * grid_size
        food_y = round(random.randrange(0, height - food_size) / grid_size) * grid_size
        snake_length += 1
        snake_speed += 1  # 먹이를 먹을 때마다 속도 증가

    # 화면을 흰색으로 채우기
    screen.fill(white)

    # 격자 그리기
    for x in range(0, width, grid_size):
        pygame.draw.line(screen, gray, (x, 0), (x, height))
    for y in range(0, height, grid_size):
        pygame.draw.line(screen, gray, (0, y), (width, y))

    # 먹이 그리기
    screen.blit(food_image, (food_x, food_y))

    # 뱀 길이 관리
    snake_head = []
    snake_head.append(snake_x)
    snake_head.append(snake_y)
    snake_list.append(snake_head)

    if len(snake_list) > snake_length:
        del snake_list[0]

    # 뱀 그리기
    for segment in snake_list:
        screen.blit(snake_image, (segment[0], segment[1]))

    # 자신의 꼬리에 부딪혔을 때 처리
    for segment in snake_list[:-1]:
        if segment == snake_head:
            pygame.quit()
            sys.exit()

    # 화면 업데이트
    pygame.display.update()

    # 초당 프레임 수 조절
    pygame.time.Clock().tick(snake_speed)
```
