```python
########################################
## 'K-mooc 이미지 처리 pillow.pdf 의 4페이지 변형
##  ---  이미지 폴더위치 추가 및 예외처리
## 이미지 열기 
## open 파일 없을시 예외처리
########################################
from PIL import Image
try:
    baseFolder='./img/'
    file= baseFolder + "flower01.jpg"
    im = Image.open(file)
    
    file=baseFolder+"flower05.jpg"
    
    im01 = Image.open(file)
    print(im.format, im.size, im.mode)
    print(im01.format, im01.size, im01.mode)
    
# IOError: 없는 파일을 열었을 때 어떻게 처리하는지 알아보기    
except IOError as err:  
    print("%s unable to load image"%file)
    
```

```python
########################################
## 이미지 열기
## 'imgList.csv' 파일에 있는 이미지 읽기 자료 리스트 읽기
## open 파일 없을시 예외처리
########################################
def imgOpen(file):
    from PIL import Image
    try:
        baseFolder='./img/'
        file= baseFolder + file
        im = Image.open(file)
        print(im.format, im.size, im.mode)
                
    # IOError: 없는 파일을 열었을 때 어떻게 처리하는지 알아보기    
    except IOError as err:  
        print("%s unable to load image"%file)
    
def prn():
    print('-'*50)
    
import csv
with open('imgList.csv', newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter='\t')
    for row in csvreader:
            prn()
            print(row)
            
            imgOpen(row[0])

```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# 이미지 열기
image_path = "./img/flower04.jpg"
img = Image.open(image_path)

# PIL 이미지를 NumPy 배열로 변환
img_array = np.array(img)

# RGB 중에서 B (파란색)을 0으로 설정
img_array[:, :, 1] = 0

# NumPy 배열을 PIL 이미지로 변환
modified_img = Image.fromarray(img_array)

# 이미지 표시
plt.imshow(modified_img)
plt.show()
```
