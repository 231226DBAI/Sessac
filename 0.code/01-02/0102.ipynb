```python
def mySum(*a):
    sum = 0
    print(type(a))
    for i in a:
        sum+=i
    return(sum)

print(mySum(1,2,3,4))


def mySum(a):
    sum = 0
    print(type(a))
    for i in a:
        sum+=i
    return(sum)

print(mySum([1,2]))
```
